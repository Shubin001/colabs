```python
import pandas as pd

# Чтение данных из файла
file_path = 'МЭП_по2022_госстат_вход.xlsx'
output_file_path='Индекс производства продукции сх_г_г_мес.xlsx'
sheet="1.3"
parametr_name="1.3. Индекс производства продукции сельского хозяйства  % к соответствующему периоду предыдущего года "
df = pd.read_excel(file_path,sheet_name=sheet, header=1)
df
```

```python
import pandas as pd
# Инициализация результирующего DataFrame
monthly_data = []

# Преобразование таблицы во временной ряд с шагом 1 месяц
for index, row in df.iterrows():
    year = row[0]
    for month, value in enumerate(row[4:], start=1):
        date = pd.Timestamp(int(year), month, 1)
        monthly_data.append([date, value])

# Создание нового DataFrame
df_monthly = pd.DataFrame(monthly_data, columns=['Дата', parametr_name])

# Запись нового листа в файл

with pd.ExcelWriter(output_file_path, engine='openpyxl', mode='w') as writer:
    df_monthly.to_excel(writer, sheet_name='Лист1', index=False)

print("Данные успешно преобразованы во временной ряд с шагом 1 месяц и сохранены в файле", output_file_path)
df_monthly
```
